### Интерполяция многочленом Ньютона

```python
def interpolation(c, x, x0):
    """
    Вычисление полинома Ньютона в точке x0.
    """
    # Степень полинома:
    n = len(x) - 1
    y0 = c[n]
    for k in range(1, n+1):
        y0 = c[n-k] + (x0 - x[n-k]) * y0
    return y0
```

```python
def coef(x, y):
    """
    Вычисление коэффициентов полинома Ньютона.
    """
    # Количество узловых точек:
    m = len(x)
    c = y.copy()
    for k in range(1, m):
        c[k:m] = (c[k:m] - c[k-1])/(x[k:m] - x[k-1])
    return c
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
plt.style.use('bmh')
%matplotlib inline
```

```python
def f(x):
    y = (1. - x) / (1 + x**2)
    return y
```

```python
a = -1.
b = 2.
x1 = np.linspace(a, b, 200)
y1 = f(x1)
plt.plot(x1, y1)
n = 3
x = np.linspace(a, b, n + 1)
y = f(x)
print(y)
plt.scatter(x, y, marker='o')
c = coef(x, y)
print(c)
y1 = interpolation(c, x, x1)
s1 = 'n=' + str(n)
sg = '--'
plt.plot(x1, y1, sg, label=s1)
plt.xlabel('x')
plt.grid(True)
plt.legend(loc=0)
plt.show()
```
